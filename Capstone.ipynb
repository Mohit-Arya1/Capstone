<b> Web Scraping Wikipedia Tables using BeautifulSoup </b>

```python
capstone_data = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
capstone_data.head()
```

```python
capstone_data.columns = ['PostalCode','Borough','Neighborhood']
```

```python
capstone_data.head()

```

```python
capstone_data.drop(capstone_data[capstone_data['Borough']=="Not assigned"].index,axis=0, inplace=True)
```

```python
capstone_data.head()
```

```python
capstone_data.shape
```

```python
capstone_data.head(19)
```

```python
capstone_data['Neighborhood'].mask(capstone_data['Neighborhood']=='Not assigned', capstone_data['Borough'], inplace=True)
```

```python
temp = capstone_data.copy()
```

```python
temp.head(10)
```

```python
tempe= temp.groupby('PostalCode').agg(lambda x: ','.join(x))
```

```python
tempe
```

```python
tempe['Borough']= tempe['Borough'].str.replace('nan|[{}\s]','').str.split(',').apply(set).str.join(',').str.strip(',').str.replace(",{2,}",",")
```
